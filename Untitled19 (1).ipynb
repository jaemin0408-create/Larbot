```python
#https://www.kaggle.com/mohamedahmedae/utkface-v2-gender-97-race-94-age-mae-0-05 코드 참고
# 필요한 라이브러리 IMPORT
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
from glob import glob
import seaborn as sns
import plotly.graph_objects as go


# 숫자로 표현되어있는 인종, 성별 데이터를 문자열로 변환해주기 위한 dictionary 생성
dataset_dict = {
    'race_id': {
        0: 'white',
        1: 'black',
        2: 'asian',
        3: 'indian',
        4: 'others'
    },
    'gender_id': {
        0: 'male',
        1: 'female'
    }
}

dataset_dict['gender_alias'] = dict((g, i) for i, g in dataset_dict['gender_id'].items())
dataset_dict['race_alias'] = dict((g, i) for i, g in dataset_dict['race_id'].items())

folder_name = 'UTKFace'

# 데이터 프레임으로 만들어주기 위한 함수 지정

def parse_dataset(dataset_path, ext='jpg'):

    def parse_info_from_file(path):
        try:
            filename = os.path.split(path)[1]
            filename = os.path.splitext(filename)[0]
            age, gender, race, _ = filename.split('_')

            return int(age), dataset_dict['gender_id'][int(gender)], dataset_dict['race_id'][int(race)]
        except Exception as ex:
            return None, None, None

    files = glob(os.path.join(dataset_path, "*.%s" % ext))

    records = []
    for file in files:
        info = parse_info_from_file(file)
        records.append(info)

    df = pd.DataFrame(records)
    df['file'] = files
    df.columns = ['age', 'gender', 'race', 'file']
    df = df.dropna()

    return df

df = parse_dataset(folder_name)

df.head()
```

```python
from tensorflow.keras.models import Model, Sequential, load_model
from tensorflow.keras.layers import BatchNormalization, Conv2D, MaxPool2D, Activation, Dropout, Lambda, Dense, Flatten, Input
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.model_selection import train_test_split
from tensorflow.keras import backend as k
```

```python
images = []
age = []
gender = []
for img in os.listdir(folder_name):
  ages = img.split("_")[0]
  genders = img.split("_")[1]
  img = cv2.imread(str(folder_name)+"/"+str(img))
  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  img = img
  images.append(np.array(img))
  age.append(np.array(ages))
  gender.append(np.array(genders))

age = np.array(age,dtype=np.int64)
images = np.array(images)
gender = np.array(gender,np.uint64)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

N = len(images)
idx = np.arange(N)

# 60% train, 20% val, 20% test
idx_train, idx_tmp = train_test_split(idx, random_state=42, test_size=0.4)
idx_val, idx_test = train_test_split(idx_tmp, random_state=42, test_size=0.5)

# images는 한 번만 나눔 (주의: fancy indexing은 복사본을 만들 수 있음)
x_train = images[idx_train]
x_valid = images[idx_val]
x_test  = images[idx_test]

# 라벨은 각각 따로 만들기
y_train_age,   y_valid_age,   y_test_age   = age[idx_train],   age[idx_val],   age[idx_test]
y_train_gender,y_valid_gender,y_test_gender= gender[idx_train],gender[idx_val],gender[idx_test]
```

```python
# Callbacks definition
callbacks_age = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint(filepath='./age-model-{epoch:02d}-{val_loss:.2f}.h5', monitor='val_loss', save_best_only=True, mode='min')
]

callbacks_gender = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ModelCheckpoint(filepath='./gender-model-{epoch:02d}-{val_loss:.2f}.h5', monitor='val_loss', save_best_only=True, mode='min')
]

# Assuming images are 200x200x3 (based on 'images' variable in kernel state)
input_shape = (200, 200, 3)

# Define age model
age_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    MaxPool2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPool2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='linear', name='age_output') # Age is regression
])
age_model.compile(optimizer='adam', loss='mae', metrics=['mae']) # MAE for regression

# age_model 학습
history_age = age_model.fit(
    x_train, y_train_age,
    validation_data=(x_valid, y_valid_age),
    batch_size=8, epochs=50, callbacks=callbacks_age
)

# (중요) age 학습 끝나면 메모리 정리 한 번 해주면 더 안전
import gc, tensorflow as tf
gc.collect()
tf.keras.backend.clear_session()

# gender_model은 clear_session() 이후에 다시 "정의부터" 만드는 게 안전함
# Define gender model (after clearing session)
gender_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    MaxPool2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPool2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid', name='gender_output') # Gender is binary classification
])
gender_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # Binary crossentropy for binary classification

history_gender = gender_model.fit(
    x_train, y_train_gender,
    validation_data=(x_valid, y_valid_gender),
    batch_size=8, epochs=50, callbacks=callbacks_gender
)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# model load (추론용으로 compile=False 권장)
age_model_path = "./age_model.h5"
gender_model_path = "./gender_model.h5"

age_model = load_model(age_model_path, compile=False)
gender_model = load_model(gender_model_path, compile=False)

# face detector
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

image_size = 200

pic = cv2.imread("./son_heung_min.jpg")
gray = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, scaleFactor=1.11, minNeighbors=8)

for (x, y, w, h) in faces:
    roi = pic[y:y+h, x:x+w]
    roi = cv2.resize(roi, (image_size, image_size))

    # 전처리 (RGB + normalize)  ※학습 때 전처리와 맞추는 게 가장 중요
    roi_rgb = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
    roi_rgb = roi_rgb.astype("float32") / 255.0
    inp = np.expand_dims(roi_rgb, axis=0)

    age_pred = float(age_model.predict(inp)[0][0])
    gender_prob = float(gender_model.predict(inp)[0][0])
    gend = "Man" if gender_prob < 0.5 else "Woman"

    # draw
    cv2.rectangle(pic, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(pic, f"Age: {age_pred:.1f} / {gend}", (x, y-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

# show
pic1 = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8, 8))
plt.imshow(pic1)
plt.axis("off")
plt.show()
```
